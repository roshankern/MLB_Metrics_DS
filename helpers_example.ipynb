```python
import matplotlib.pyplot as plt
import seaborn as sns

import mlb_metrics_helpers

import warnings
warnings.filterwarnings("ignore")
```

```python
# get pitcher data for balls crossing the plates
aardsma_id = mlb_metrics_helpers.player_id("Aardsma", "David")
aardsma_general_stats = mlb_metrics_helpers.player_general_metrics(
    aardsma_id, timeline_type="career"
)
aardsma_timeline = mlb_metrics_helpers.parse_career_timeline(aardsma_general_stats)
aardsma_specific_stats = mlb_metrics_helpers.player_specific_metrics(
    aardsma_id,
    metric_type="pitching",
    start_dt=aardsma_timeline[0],
    end_dt=aardsma_timeline[1],
)
aardsma_plate_crossing_metrics = mlb_metrics_helpers.plate_crossing_metrics(aardsma_specific_stats)

aardsma_plate_crossing_metrics
```

```python
# get location data for different pitches crossing the plate
pitch_name_locations = aardsma_plate_crossing_metrics[['pitch_name', 'plate_x', 'plate_z']]

# Set up the plot
plt.figure(figsize=(8, 6))

# Create scatter plot
sns.scatterplot(data=pitch_name_locations, x="plate_x", y="plate_z", hue="pitch_name", s=100, alpha=0.5)

# Customize the axes
plt.axvline(x=-0.71, color='gray', linestyle='--')  # left bound of strike zone (17 inches / 2)
plt.axvline(x=0.71, color='gray', linestyle='--')  # right bound of strike zone (17 inches / 2)

# Add labels and legend
plt.title("Scatter Plot of Types of Pitches Crossing Plate")
plt.xlabel("Horizontal Position (feet)")
plt.ylabel("Vertical Position (feet)")
plt.legend(title="Pitch Name")

# Show plot
plt.show()
```

```python
# get batter data
carroll_id = mlb_metrics_helpers.player_id("Carroll", "Corbin")
carroll_general_stats = mlb_metrics_helpers.player_general_metrics(carroll_id, timeline_type="career")
carroll_timeline = mlb_metrics_helpers.parse_career_timeline(carroll_general_stats)

carroll_specific_stats = mlb_metrics_helpers.player_specific_metrics(
    carroll_id,
    metric_type="batting",
    start_dt=carroll_timeline[0],
    end_dt=carroll_timeline[1],
)
carroll_plate_crossing_metrics = mlb_metrics_helpers.plate_crossing_metrics(carroll_specific_stats)

carroll_plate_crossing_metrics
```

```python
# get location data for batting events crossing the plate
description_locations = carroll_plate_crossing_metrics[['description', 'plate_x', 'plate_z']]

# Set up the plot
plt.figure(figsize=(8, 6))

# Create scatter plot
sns.scatterplot(data=description_locations, x="plate_x", y="plate_z", hue="description", s=100, alpha=0.3)

# Customize the axes
plt.axvline(x=-0.71, color='gray', linestyle='--')  # left bound of strike zone (17 inches / 2)
plt.axvline(x=0.71, color='gray', linestyle='--')  # right bound of strike zone (17 inches / 2)

# Add labels and legend
plt.title("Scatter Plot of Batting Events Crossing Plate")
plt.xlabel("Horizontal Position (feet)")
plt.ylabel("Vertical Position (feet)")
plt.legend(title="Pitch Name")

# Show plot
plt.show()
```
